```python
import numpy as np
import pandas as pd
```

```python
def loading_txt(root, first_row, first_column):
    #read txt method one
    f = open(root)
    line = f.readline()
    locations = []

    while line:
        line = f.readline().strip('\n').split('|')  
        if len(line) == 1:
            break
        locations.append(line[first_column:])
    f.close()
    return locations
```

```python
df_locations = pd.DataFrame(test_locations)
df_locations
```

```python
df_prediction = pd.DataFrame(prediction_values)
df_prediction[0] = df_prediction[0].str.split("/", expand = True).iloc[:,-1]
df_prediction
```

```python
#sort by name
#Don't sort original images, because the sort is correct
if df_prediction[0][0] != 'img_fr_nov15_000.000.png':     
    df_prediction[0] = df_prediction[0].apply(lambda x : ('img0' + x[-6:-4]) if x[-7] == '_' else ('img'+x[-7:-4]))
    df_prediction = df_prediction.sort_values(by= 0)
    df_prediction = df_prediction.reset_index(drop=True)
```

```python
#change dtypes
for i in range(1,df_locations.shape[1]):
    df_locations[i] = df_locations[i].astype(np.float32)
for i in range(1,df_prediction.shape[1]):
    df_prediction[i] = df_prediction[i].astype(np.float32)
```

```python
df_locations.columns = ['filename','img_width','img_height','x1','y1','x2','y2','center_x','center_y','width','height']
df_locations
```

```python
df_merge = pd.merge(df_locations.loc[:, ['filename','center_x','center_y','width']], df_prediction.loc[:, :4], on = df_locations.index)
df_merge.rename(columns={'center_x': 'x','center_y': 'y', 1:'xp', 2:'yp', 3:'rp'}, inplace=True)
df_merge['r'] = df_merge['width']/2
df_merge = df_merge[['filename', 'x', 'y', 'r', 'xp', 'yp', 'rp']]
df_merge
```

```python
def calculate_overlap(x):
    dis = x['dis']
    pi = np.arccos(-1.0)
    if (x['r'] + x['rp']) <= dis:
        return 0
    if (x['r'] - x['rp']) >= dis:
        return  pi * x['rp'] * x['rp']
    if (x['rp'] - x['r']) >= dis:
        return  pi * x['r'] * x['r']
    angle1 = np.arccos((x['r'] * x['r'] + dis * dis - x['rp'] * x['rp']) / (2 * dis * x['r']))
    angle2 = np.arccos((x['rp'] * x['rp'] + dis * dis - x['r'] * x['rp']) / (2 * dis * x['rp']))
    return x['r'] * x['r'] * angle1 + x['rp'] * x['rp'] * angle2 - np.sin(angle1) * x['r'] * dis
#     print(x)
# df_merge.iloc[:10, :].apply(calculate_overlap, axis=1)
```

```python
df_merge['dis'] = np.sqrt(np.square(df_merge['x'] - df_merge['xp']) + np.sqrt(np.square(df_merge['y'] - df_merge['yp'])))
df_merge['overlap_area'] = df_merge.apply(calculate_overlap, axis=1)
df_merge['original_circle_area'] = np.arccos(-1.0) * np.square(df_merge['r'])
df_merge['predicted_circle_area'] = np.arccos(-1.0) * np.square(df_merge['rp'])
acc = ((df_merge['overlap_area']/df_merge['original_circle_area'] > 0.5) & 
 (df_merge['predicted_circle_area']/df_merge['original_circle_area'] < 1.5)).sum()/df_merge.shape[0]

print('acc is: ', acc*100, '%')
```
